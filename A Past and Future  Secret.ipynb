```python
# integration
```

```python
discography = pd.concat([tales, somewhere, imaginations, forgotten, nightfall], ignore_index = True)
discography
```

```python
discography.drop('No.', axis = 1, inplace = True)
discography
```

```python
discography.info()
```

```python
# Here I amost dropped the project, as Length has been a friggin object from the beginning...
# had to look this up and used the solution here: https://stackoverflow.com/questions/47797290/convert-hhmm-to-float-pandas
# I think we covered it in class (?)
```

```python
y = discography.Length.str.split(':')
i, j = y.str[0], y.str[1]

discography['duration'] = i.astype(int) + (j.astype(float) / 60)
```

```python
# Ending comments
# Duration of songs trended downwards
# Their most kick-ass albums (in my view) are Somewhere, Imaginations, with medium-to-high length of songs
# The only pure concept album (Nightfall) has a large spread re: duration, which is explainable
# as it contains role-playing interludes, narration etc. It is also one of their most popular (I think).
```

```python
# encore
```

```python
discography['year'] = discography['year'].astype(int)
```

```python
discography.info()
```

```python
def period(year):
  if year < 1995:
    return 'early 90s'
  else:
    return 'late 90s'
```

```python
discography['phase'] = discography['year'].apply(period)
discography['phase']
```

```python
discography.head()
```

```python
sns.violinplot(data = discography, y = 'duration', hue = 'phase')
```
